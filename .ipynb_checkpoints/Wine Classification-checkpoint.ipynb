# 2. Veri Seti Kalite Kontrolleri

Veri setindeki eksik değerlerin tespiti için her sütun kontrol edilmiştir.

```python
print("\n2.1 Eksik Değer Analizi:")
missing = df.isnull().sum()
print(missing)
print(f"Toplam eksik değer: {missing.sum()}")
```

Veri setinde hiç eksik değer bulunmamaktadır. Bu durum, veri setinin temiz olduğunu göstermektedir. Eksik değer doldurma işlemine gerek kalmamıştır.

```python
print("\n2.2 Aykırı Değer Analizi (IQR Yöntemi):")
def detect_outliers_iqr(df_temp):
    outlier_indices = []
    numeric_cols = df_temp.select_dtypes(include=[np.number]).columns
    for col in numeric_cols:
        Q1 = df_temp[col].quantile(0.25)
        Q3 = df_temp[col].quantile(0.75)
        IQR = Q3 - Q1
        lower = Q1 - 1.5 * IQR
        upper = Q3 + 1.5 * IQR
        outlier_idx = df_temp[(df_temp[col] < lower) | (df_temp[col] > upper)].index
        if len(outlier_idx) > 0:
            print(f'Sütun [{col}] - {len(outlier_idx)} adet aykırı değer bulundu.')
        outlier_indices.extend(outlier_idx)
    return list(set(outlier_indices))
```

```python
outlier_indices = detect_outliers_iqr(df.drop('target', axis=1))
print(f"Aykırı değer sayısı: {len(outlier_indices)} ({len(outlier_indices)/len(df)*100:.2f}%)")
```

Aykırı değerlerin tespiti için IQR (Interquartile Range) yöntemi kullanılmıştır. Toplam 17 gözlem aykırı değer içermektedir. Bu, veri setinin \%9.55'ine karşılık gelmektedir.

Aykırı değerlerin etkileri şunlardır:
- **Olumlu Etkiler**: Aykırı değerler doğal varyasyonu temsil edebilir ve modelin genelleme yeteneğini artırabilir.
- **Olumsuz Etkiler**: Doğrusal modelleri (Lojistik Regresyon) olumsuz etkileyebilir. Uzaklık tabanlı algoritmaları (KNN) bozabilir. Model performansını düşürebilir.

Bu çalışmada aykırı değerler çıkarılmayacak, ancak robust scaler kullanılacaktır.

```python
print("\n6.1 PCA (Principal Component Analysis):")
pca = PCA()
pca.fit(X_train)
exp_var = pca.explained_variance_ratio_
cum_var = np.cumsum(exp_var)
mean_var = exp_var.mean()
n_components_pca = np.argmax(cum_var >= cum_var[-1] * 0.95) + 1

print(f"PCA bileşen sayısı (95% varyans): {n_components_pca}")
```

```python
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 4))
ax1.plot(range(1, len(exp_var)+1), exp_var, 'bo-')
ax1.set_xlabel('PCA Bileşen Sayısı')
ax1.set_ylabel('Explained Variance Ratio')
ax1.set_title('PCA Explained Variance')
ax1.grid()

ax2.plot(range(1, len(cum_var)+1), cum_var, 'ro-')
ax2.axhline(y=0.95, color='g', linestyle='--', label='95% Threshold')
ax2.set_xlabel('PCA Bileşen Sayısı')
ax2.set_ylabel('Cumulative Explained Variance')
ax2.set_title('PCA Cumulative Explained Variance')
ax2.legend()
ax2.grid()
plt.tight_layout()
plt.savefig('grafikler/pca_variance.png', dpi=100)
plt.show()
plt.close()
```

# 7. Modellerin Kurulması ve Validasyon

```python
def get_algorithm_instance(algo_name):
    """Her veri versiyonu için yeni model instance'ı oluştur"""
    if algo_name == 'Logistic Regression':
        return LogisticRegression(max_iter=1000, random_state=42)
    elif algo_name == 'Decision Tree':
        return DecisionTreeClassifier(random_state=42)
    elif algo_name == 'Random Forest':
        return RandomForestClassifier(n_estimators=100, random_state=42)
    elif algo_name == 'XGBoost':
        return XGBClassifier(objective='multi:softprob', num_class=3, random_state=42, verbose=0)
    elif algo_name == 'Naive Bayes':
        return GaussianNB()
```

```python
algorithm_names = ['Logistic Regression', 'Decision Tree', 'Random Forest', 'XGBoost', 'Naive Bayes']
```

```python
data_versions = {
    'Ham Veri': (X_train, X_val, X_test),
    'PCA': (X_train_pca, X_val_pca, X_test_pca),
    'LDA': (X_train_lda, X_val_lda, X_test_lda)
}
```

```python
results = []
models_dict = {}

for data_name, (X_tr, X_v, X_te) in data_versions.items():
    for algo_name in algorithm_names:
        model = get_algorithm_instance(algo_name)
        model.fit(X_tr, y_train)
        y_pred = model.predict(X_v)
        
        y_pred_proba = model.predict_proba(X_v)

        try:
            if len(np.unique(y)) == 2:
                roc_auc = roc_auc_score(y_val, y_pred_proba[:, 1])
            else:
                roc_auc = roc_auc_score(y_val, y_pred_proba, 
                                        multi_class='ovr', 
                                        average='weighted')
        except Exception as e:
            print(f"ROC-AUC hatası ({data_name} - {algo_name}): {e}")
            roc_auc = np.nan

        results.append({
            'Veri': data_name,
            'Algoritma': algo_name,
            'Accuracy': accuracy_score(y_val, y_pred),
            'Precision': precision_score(y_val, y_pred, average='macro', zero_division=0),
            'Recall': recall_score(y_val, y_pred, average='macro', zero_division=0),
            'F1': f1_score(y_val, y_pred, average='macro', zero_division=0),
            'ROC-AUC': round(roc_auc, 4)
        })
        
        models_dict[f"{data_name}_{algo_name}"] = model
```

```python
results_df = pd.DataFrame(results)

results_df = results_df.sort_values(by='F1', ascending=False).reset_index(drop=True)

results_df = results_df.round({
    'Accuracy': 4,
    'Precision': 4,
    'Recall': 4,
    'F1': 4,
    'ROC-AUC': 4
})

print("\nTüm Modellerin Validation Performansı (F1 Skoruna Göre Sıralı):")
print("="*100)
print(results_df.to_string(index=False))
```

Tabloya göre birden fazla model %100’e yakın performans sergilemiştir. En düşük performansı PCA veri temsili ile XGBoost modeli vermiştir.

```python
best_idx = 2 # results_df['F1'].idxmax()
best_result = results_df.loc[best_idx]
print(f"\nEn İyi Model: {best_result['Veri']} - {best_result['Algoritma']}")
print(f"Accuracy: {best_result['Accuracy']:.4f}")
```

```python
y_test_bin = label_binarize(y_test, classes=[0, 1, 2])
n_classes = y_test_bin.shape[1]

y_pred = best_model.predict(X_test)
y_pred_proba = best_model.predict_proba(X_test)

plt.figure(figsize=(15, 12))
colors = ['blue', 'red', 'green']
class_names = ['Sınıf 0', 'Sınıf 1', 'Sınıf 2']

for i in range(3):
    fpr, tpr, thresholds = roc_curve(y_test_bin[:, i], y_pred_proba[:, i])
    
    plt.subplot(3, 2, i*2 + 1)
    plt.plot(thresholds, tpr, 'g-', label='True Positive Rate (TPR)', linewidth=2)
    plt.plot(thresholds, fpr, 'r-', label='False Positive Rate (FPR)', linewidth=2)
    plt.axvline(x=0.5, color='gray', linestyle='--', alpha=0.7, label='Threshold = 0.5')
    
    plt.xlabel('Threshold Değeri')
    plt.ylabel('Oran')
    plt.title(f'{class_names[i]} - Threshold Etkisi')
    plt.legend()
    plt.grid(alpha=0.3)
    plt.xlim([0, 1])
    plt.ylim([0, 1.1])
    
    plt.subplot(3, 2, i*2 + 2)
    
    precisions = []
    recalls = []
    for threshold in np.linspace(0, 1, 100):
        y_pred_binary = (y_pred_proba[:, i] >= threshold).astype(int)
        precision = precision_score(y_test_bin[:, i], y_pred_binary, zero_division=0)
        recall = recall_score(y_test_bin[:, i], y_pred_binary, zero_division=0)
        precisions.append(precision)
        recalls.append(recall)
    
    plt.plot(recalls, precisions, 'b-', linewidth=2)
    plt.xlabel('Recall')
    plt.ylabel('Precision')
    plt.title(f'{class_names[i]} - Precision-Recall Eğrisi')
    plt.grid(alpha=0.3)
    plt.xlim([0, 1.1])
    plt.ylim([0, 1.1])

plt.tight_layout()
plt.savefig('grafikler/class_threshold_analysis.png', dpi=100)
plt.show()
```

Sınıf 0, threshold 0.7, Sınıf 1 için threshold 0.8, Sınıf 2 için threshold 0.9’dur. Threshold arttıkça TPR sabit kalıyor ve FPR düşüyor. Bu, modelin pozitif sınıfı ayırt etme gücünün çok yüksek olduğunu gösterir. Negatifleri yanlış pozitive çevirme eğilimi threshold yükseldikçe tamamen kayboluyor. Daha sıkı eşik bile pozitifleri kaçırmıyor. Negatif yanlışlar azalıyor
